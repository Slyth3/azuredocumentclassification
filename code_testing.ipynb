```python
from azure.storage.blob import BlobServiceClient
import json
import time
from requests import get, post
import os
import requests
from collections import OrderedDict
from pathlib import Path
import base64
# import pandas as pd
```

```python
def load_local_settings():
    settings_path = Path("local.settings.json")

    if not settings_path.exists():
        return  

    with open(settings_path) as f:
        values = json.load(f).get("Values", {})

    for key, value in values.items():
        os.environ.setdefault(key, value)
load_local_settings()
```

```python
wait_sec = 10
modelId = "prebuilt-layout"

apim_key = os.getenv("docintelligencekey")
endpoint = os.getenv("docintelligenceendpoint")
```

```python
pdf_path = os.getenv("pdf_path_example") 
with open(pdf_path, "rb") as f:
    pdf_bytes = f.read()
print(pdf_bytes)

#added 
pdf_base64 = base64.b64encode(pdf_bytes).decode("utf-8")
pdf_base64
```

## Code

```python
post_url = f"{endpoint}/documentintelligence/documentModels/{modelId}:analyze?api-version=2024-11-30"

headers = {
    "Ocp-Apim-Subscription-Key": apim_key,
    "Content-Type": "application/json"
} 

payload = {
    "base64Source": pdf_base64
}

resp = requests.post(url=post_url, json=payload, headers=headers)
print(resp.status_code)
print(resp.headers)
print(resp.text)
```

```python
if resp.status_code != 202:
    print("POST analyze failed:\n%s" % resp.text)
    quit()
print("POST analyze succeeded:\n%s" % resp.headers)

get_url = resp.headers["operation-location"]

# The layout API is async therefore the wait statement
time.sleep(wait_sec)
resp = requests.get(url=get_url, headers={"Ocp-Apim-Subscription-Key": apim_key})

resp_json = json.loads(resp.text)
```

```python
status = resp_json["status"]

if status == "succeeded":
    print("POST Layout Analysis succeeded:\n%s")
    results = resp_json
else:
    print("GET Layout results failed:\n%s")
    quit()

results = resp_json
```

# Ai Foundry

```python
document_types = "Income Certificate, Medical Aid Tax Certificate, Retirement Annuity Certificate, Investment Income Certificate, Medical Expenses, Travel Log Book, Other"

system_prompt = f"""You are a document classification agent for the South African Revenue Service. 
Your task is to analyze the provided text and classify it into one or more document types. 
Valid types are strictly limited to: {document_types}. 

If 'Other' is used, you must specify the exact type in plain text (e.g., 'Proof of donations', 'travel logbook', 'Bank Statement', 'Invoice' etc). 
You must respond ONLY in valid JSON with the following structure:"
{{
  "Result": ["<list of identified types>"],
  "Explanation": "<short explanation of why these types were chosen>
}}

Do not include any text outside of this JSON object. If more than one type is identified, separate them with commas."""

user_prompt = f"""Analyze the following text and classify the document type(s)"""
```

```python
import os
from openai import AzureOpenAI

openai_endpoint = os.getenv("openai_endpoint")
model_name = "gpt-4.1"
openai_key = os.getenv("openai_key")
api_version = "2024-12-01-preview"

client = AzureOpenAI(
    api_version=api_version,
    azure_endpoint=openai_endpoint,
    api_key=openai_key,
)

completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": system_prompt,
        },
        {
            "role": "user",
            "content": user_prompt + f"\n\nExtracted Text:\n{content}",
        }
    ],
    max_completion_tokens=32768,
    temperature=0.0,
    response_format={"type": "json_object"},
    model=model_name
)

print(completion.choices[0].message.content)
```

```python
import json
from azure.eventhub import EventHubProducerClient, EventData
from azure.identity import DefaultAzureCredential
from datetime import datetime
import os


def eventhub_save(payload: dict):
        event = {
            "result": payload["result"],
            "explanation": payload["explanation"]
        }

        credential = DefaultAzureCredential()

        producer = EventHubProducerClient(
            fully_qualified_namespace=os.getenv("Eventhub_endpoint"),
            eventhub_name=os.getenv("Eventhub_name"),
            credential=credential
        )

        with producer:
            batch = producer.create_batch()
            batch.add(EventData(json.dumps(event)))
            producer.send_batch(batch)

        return "Event sent to Fabric Eventstream"

# eventhub_save(result)
```
